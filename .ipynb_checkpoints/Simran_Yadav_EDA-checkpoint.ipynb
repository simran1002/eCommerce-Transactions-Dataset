Task 1: Exploratory Data Analysis (EDA) and Business Insights

Step 1: Load and Inspect Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
customers = pd.read_csv('Customers.csv')
products = pd.read_csv('Products.csv')
transactions = pd.read_csv('Transactions.csv')

# Display first few rows
print(customers.head())
print(products.head())
print(transactions.head())

# Check for missing values and duplicates
print(customers.isnull().sum())
print(products.isnull().sum())
print(transactions.isnull().sum())
print(customers.duplicated().sum())
print(products.duplicated().sum())
print(transactions.duplicated().sum())

```

Step 2: Summary Statistics

```python
# Summary statistics for numerical columns
print(customers.describe())
print(products.describe())
print(transactions.describe())

```

Step 3: Visualizations

```python
# Plot distribution of 'Price', 'Quantity', and 'TotalValue'
plt.figure(figsize=(12, 6))
sns.histplot(transactions['Price'], kde=True, bins=30)
plt.title('Price Distribution')
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(transactions['Quantity'], kde=True, bins=30)
plt.title('Quantity Distribution')
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(transactions['TotalValue'], kde=True, bins=30)
plt.title('Total Value Distribution')
plt.show()

# Visualize product categories
plt.figure(figsize=(12, 6))
sns.countplot(x='Category', data=products)
plt.title('Product Category Distribution')
plt.xticks(rotation=90)
plt.show()

# Time series of transactions
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'])
transactions.groupby(transactions['TransactionDate'].dt.to_period('M')).size().plot()
plt.title('Transactions Over Time')
plt.show()
```
